```r
summaryRprof()
#these functions took 1.32 seconds and we can see the call stack measured in the tables
```

```r
summaryRprof()$by.total
#this tells us how mucn time is spent in each of these functions 
#we can see read_csv spent 1.30 seconds and 98.48% of the time was spent in this function
```

```r
summaryRprof()$by.self
#this does the same, but subtracts out all lower level function calls 
#the main function call took 1.30 seconds and .escape_vec took .02 seconds (probably to exit the function)
```

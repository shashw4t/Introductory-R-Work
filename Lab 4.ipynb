```r
#page 13
numExp = 1500
tosses = 6
exp = rep(0, numExp)
for (i in 1:numExp){
    flips = sample(c('H', 'T'), tosses, replace = T)
    exp[i] = sum(flips == 'T')
}
hist(exp, breaks =-.5:6.5, main = 'Experimental Data vs. Binomial Distribution')
lines(0:6, dbinom(0:6, 6, .5)*numExp)
points(0:6, dbinom(0:6, 6, .5)*numExp)
```

```r
#page 17 
pbinom(65, 100, 0.7, lower.tail = F)
1-pbinom(65, 100, 0.7)
```

```r
#page 23 
x = rnorm(2000)
heightCalc = hist(x, plot = F)
yAxis = range(0, heightCalc$density, dnorm(0))
yAxis
hist(x, freq = F, ylim = yAxis, main = 'Normal Distribution Histogram')
curve(dnorm(x), add=T)
```
